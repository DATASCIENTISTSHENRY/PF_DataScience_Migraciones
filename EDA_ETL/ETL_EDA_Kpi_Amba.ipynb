## Creación de tabla con Totales

```python
## Se crea una función para agregar columna año y fecha a cada dataset
## se reordenan las columnas y se utlizan solo las necesarias 
# df sera el dataframe que se le ingresa
# año sera el año que nesecitamos agregar
#trimestre sera el trimestre que corresponde a ese año de la tabla 


def agregar_año_trimestre(df, año, trimestre):
    # Copiar el DataFrame para evitar modificar el original
    df = df.copy()

    # Agregar las columnas de AÑO y TRIMESTRE con los valores proporcionados
    df['AÑO'] = año
    df['TRIMESTRE'] = trimestre
    
    # Reordenar las columnas y eliminar las columnas no deseadas
    df = df[['AÑO', 'TRIMESTRE','PROVINCIA', 'TOTAL']]


    return df
```

## Creación de Tablas Anuales

```python
#Concateno dataframes para lograr un archivo por año

def concatenar_trimestres(lista_dataframes, data_id):
    """
    Concatena una lista de DataFrames correspondientes a cada trimestre en un solo DataFrame y además con la columna porvincia_id del dataframe cod_provincias.
    Convierto los nombres de las columnas a minúsculas para normalizar los datasets
    
    
    Parametro:
        - lista_dataframes (list): Lista de DataFrames a concatenar.
        - data_id (pd.Dataframe): contiene la columna provincia_id
    Retorna:
        - pd.DataFrame: DataFrame combinado.
    """
    df_combined = pd.concat(lista_dataframes, ignore_index=True)
    
    # Convertir los nombres de las columnas a minúsculas
    df_combined.columns = map(str.lower, df_combined.columns)
    
    # Agregar la columna provincia_id al DataFrame combinado
    df_combined_with_id = pd.merge(df_combined, data_id, left_on='provincia', right_on='provincia_nombre', how='left')

    return df_combined_with_id
```
