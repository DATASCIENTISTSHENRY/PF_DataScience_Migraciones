## Creación de tabla con Totales

```python
## Se crea una función para agregar columna año y fecha a cada dataset
## se reordenan las columnas y se utlizan solo las necesarias 
# df sera el dataframe que se le ingresa
# año sera el año que nesecitamos agregar
#trimestre sera el trimestre que corresponde a ese año de la tabla 


def agregar_año_trimestre(df, año, trimestre):
    # Copiar el DataFrame para evitar modificar el original
    df = df.copy()

    # Agregar las columnas de AÑO y TRIMESTRE con los valores proporcionados
    df['AÑO'] = año
    df['TRIMESTRE'] = trimestre
    
    # Reordenar las columnas y eliminar las columnas no deseadas
    df = df[['AÑO', 'TRIMESTRE','PROVINCIA', 'TOTAL']]


    return df
```

## Creación de Tablas Anuales

```python
#Concateno dataframes para lograr un archivo por año

def concatenar_trimestres(lista_dataframes, data_id):
    """
    Concatena una lista de DataFrames correspondientes a cada trimestre en un solo DataFrame y además con la columna porvincia_id del dataframe cod_provincias.
    Convierto los nombres de las columnas a minúsculas para normalizar los datasets
    
    
    Parametro:
        - lista_dataframes (list): Lista de DataFrames a concatenar.
        - data_id (pd.Dataframe): contiene la columna provincia_id
    Retorna:
        - pd.DataFrame: DataFrame combinado.
    """
    df_combined = pd.concat(lista_dataframes, ignore_index=True)
    
    # Convertir los nombres de las columnas a minúsculas
    df_combined.columns = map(str.lower, df_combined.columns)
    
    # Agregar la columna provincia_id al DataFrame combinado
    df_combined_with_id = pd.merge(df_combined, data_id, left_on='provincia', right_on='provincia_nombre', how='left')

    return df_combined_with_id
```

```python
plt.figure(figsize=(16, 8))

# Filtrar los valores sin datos o nulos
df_filtrado = df_extranjeros[df_extranjeros["provincia_nombre"] != 0]

# Obtener el orden de las provincias por la cantidad de extranjeros (de mayor a menor)
orden_provincias = df_filtrado['provincia_nombre'].value_counts().index

# Se utiliza countplot para contar la frecuencia de cada valor en la columna 'provincia_nombre'
ax = sns.countplot(data=df_filtrado, x='provincia_nombre', palette="winter", order=orden_provincias)

plt.ylabel("Frecuencia de cantidad de extranjeros")
plt.xlabel("Provincias")
plt.title("Histograma de distribución de extranjeros por provincia 2012-2022")

plt.xticks(rotation=45, ha="right")  # Rotar los nombres de las provincias para mayor legibilidad

# Agregar etiquetas a las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()  # Ajustar el diseño para evitar cortar las etiquetas
plt.show()
```

```python
# Se agrupa por la cantidad de extranjeros por provincia
data = df_extranjeros.groupby(['provincia_nombre'])['cantidad'].sum().reset_index()
data["cantidad"]=data["cantidad"].astype("Int64")

# Se ordena el DataFrame por la cantidad de extranjeros de mayor a menor
data = data.sort_values(by='cantidad', ascending=False)

# Se grafica
plt.figure(figsize=(10,4))

ax = sns.barplot(x='provincia_nombre', y='cantidad', data=data, palette="winter")
ax.set_title('Cantidad de extranjeros por provincia')
ax.set_xlabel('Provincias') ; ax.set_ylabel('Cantidad de Extranjeros millones')

# Agregar etiquetas de número de extranjeros en cada barra
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=8, color='black', xytext=(0, 5),
                textcoords='offset points')

# Se muestra el gráfico y se rotan las etiquetas 
plt.xticks(rotation=45, ha="right")
plt.show()
```

### correlación provincia y país de origen
