Crecimiento del PIB (% anual)

Tasa de crecimiento anual porcentual del PIB a precios de mercado en moneda local, a precios constantes. Los agregados están expresados en dólares de los Estados Unidos a precios constantes del año 2010. El PIB es la suma del valor agregado bruto de todos los productores residentes en la economía más todo impuesto a los productos, menos todo subsidio no incluido en el valor de los productos. Se calcula sin hacer deducciones por depreciación de bienes manufacturados o por agotamiento y degradación de recursos naturales.

```python
#Leo el dataset
df_cre_pbi = pd.read_excel('/Users/gaston/Downloads/Banco_Mundial_Crecimiento_del_PBI_%anual.xls', sheet_name='Data')
df_cre_pbi.head()
```

```python
#Borro las 2 primeras filas del df
df_cre_pbi = df_cre_pbi.iloc[2:]

# Restablezco los índices del DataFrame después de eliminar la primera fila
df_cre_pbi = df_cre_pbi.reset_index(drop=True)

# Establezco la primera fila como los nombres de las columnas
df_cre_pbi.columns = df_cre_pbi.iloc[0]

# Elimino la primera fila actual
df_cre_pbi = df_cre_pbi.iloc[1:]

# Restablezco los índices del DataFrame después de eliminar la primera fila
df_cre_pbi = df_cre_pbi.reset_index(drop=True)
df_cre_pbi.head()
```

```python
# Convierto los nombres de las columnas a cadenas (strings)
df_cre_pbi.columns = df_cre_pbi.columns.astype(str)

# Eliminar ".0" de los nombres de las columnas
df_cre_pbi.columns = df_cre_pbi.columns.str.replace('.0', '')
df_cre_pbi.head()
```

```python
# Filtro los países en el DataFrame (los 10 destinos más frecuentas a donde emigran los argentinos y argentina)
destinos_deseados = [
    "Argentina",
    "España",
    "Estados Unidos",
    "Paraguay",
    "Chile",
    "Italia",
    "Israel",
    "Bolivia",
    "Brasil",
    "Uruguay",
    "Canadá"
]

filtered_df = df_cre_pbi[df_cre_pbi['Country Name'].isin(destinos_deseados)]

# Reseteo los índices
filtered_df = filtered_df.reset_index(drop=True)

filtered_df
```
