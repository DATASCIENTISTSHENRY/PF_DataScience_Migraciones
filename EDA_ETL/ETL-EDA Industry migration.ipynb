```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Industry Migration

Migración industrial: industrias ganadas y perdidas. Basado en la industria asociada con la empresa de un miembro en el momento de la migración. La ganancia o pérdida neta de miembros de otro país que trabajan en una industria determinada dividida por la cantidad de miembros de LinkedIn que trabajan en esa industria en el país objetivo (o seleccionado), multiplicada por 10,000.

```python
#Leo el dataset
df_ind_mig = pd.read_csv('/Users/gaston/Downloads/industry_migration_public.csv', delimiter=',')
df_ind_mig.head()
```

```python
df_ind_mig = df_ind_mig.iloc[:, :13]
df_ind_mig
```

```python
#Defino un dataframe solo con Argentina
df_ind__mig_ar = df_ind_mig.loc[df_ind_mig['country_name']=='Argentina']
df_ind_mig_ar = df_ind__mig_ar.reset_index(drop=True)
df_ind__mig_ar
```

```python
df_ind_mig_ar.info()
```

Observamos que no tenemos datos nulos

```python
#idenfifico los nulos
nan_values = df_ind_mig_ar.isna()

#Cuento los NaN en cada columna.
nan_count = nan_values.sum()

# Cuento el total de nan para todo el df
total_nan_count = nan_count.sum()

print("Cantidad de valores NaN por columna:")
print(nan_count)

print("\nCantidad total de valores NaN en el DataFrame:", total_nan_count)
```

```python
df_ind_mig_ar.describe()
```

```python
# Creo un diccionario para mapear los nombres de las columnas originales a los nombres simplificados (solo el año).
column_name_mapping = {}
for column in df_ind_mig_ar.columns:
    if 'net_per_10K_' in column:
        year = column.split('_')[-1]
        column_name_mapping[column] = year

# Renombo las columnas usando el diccionario de mapeo.
df_ind_mig_ar = df_ind_mig_ar.rename(columns=column_name_mapping)

# Ahora, las columnas tendrán solo el año como nombre.
# Por ejemplo, "net_per_10K_2015" se convertirá en "2015".

df_ind_mig_ar
```

```python
#Remmplazo el código del país por el código internacional para poder unificar luego la información del proyecto
df_ind_mig_ar['country_code'] = df_ind_mig_ar['country_code'].str.replace('ar','ARG')
```

```python
#Traspongo el dataframe, de forma de dejar los años como filas y los valores del indicador en una columna con el valor para cada año
# Melt el DataFrame
df_ind_mig_ar_melted = pd.melt(df_ind_mig_ar, id_vars=["country_code", "country_name", "wb_income", "wb_region", "isic_section_index", "isic_section_name",'industry_id','industry_name'], var_name="Year", value_name="Value")

# Muestra el DataFrame resultante
df_ind_mig_ar_melted
```

```python
# Visualización de datos
df_ind_mig_ar_melted['Year'] = df_ind_mig_ar_melted['Year'].astype(int)
for year in range(2015, 2020):
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df_ind_mig_ar_melted[df_ind_mig_ar_melted['Year'] == year], x='Value')
    plt.title(f'Distribución de Valores en {year}')
    plt.show()
```

```python
data_grouped = df_ind_mig_ar_melted.groupby('Year')['Value'].describe()
data_grouped
```

```python
# Crear un gráfico de violín para mostrar la distribución de 'Value' a lo largo de los años
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_ind_mig_ar_melted, x='Year', y='Value', hue='Year', palette="Set3")
plt.title('Distribución de Valores a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Value')
plt.legend(title='Año', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

En base a estos datos, podemos observar variabilidad significativa en el valor a lo largo de los años. Los valores oscilan desde negativos (indicando pérdidas) hasta positivos (indicando ganancias) en diferentes años. Los años 2017 y 2018 muestran valores promedio más altos en comparación con los otros años. El año 2019 muestra una disminución significativa con un valor promedio negativo.

Este análisis proporciona una visión general de la variabilidad de los datos a lo largo de los años y puede ayudarte a entender mejor la tendencia económica en función de estos valores.

### industry_name

```python
# Crear un DataFrame para sumar los valores por categoría de habilidades y año
industry_sums_by_year = df_ind_mig_ar_melted.groupby(['Year', 'industry_name'])['Value'].sum().unstack()

# Mostrar la tabla
industry_sums_by_year
```

```python
# Obtener una lista de años únicos en tus datos
years = df_ind_mig_ar_melted['Year'].unique()

# Iterar a través de los años
for year in years:
    # Filtrar los datos para el año actual
    data_for_year = df_ind_mig_ar_melted[df_ind_mig_ar_melted['Year'] == year]
    
    # Calcular la suma de valores por "industry_name" para el año actual y ordenar de mayor a menor
    top_10_categories = data_for_year.groupby('industry_name')['Value'].sum().nlargest(10)
    
    # Crear un gráfico de barras horizontales para el año actual
    plt.figure(figsize=(10, 6))
    top_10_categories.plot(kind='barh')
    
    # Configurar el gráfico
    plt.title(f'Top 10 industry_name con más valores en {year}')
    plt.xlabel('Suma de Valores')
    plt.ylabel('Industry_name')
    
    # Mostrar el gráfico
    plt.show()
```

```python
# Agrupar y sumar los valores por categoría de habilidades
industry_sums = df_ind_mig_ar_melted.groupby('industry_name')['Value'].sum()

# Crear un DataFrame para mostrar el total general
industry_sums_total = pd.DataFrame({'Total General': industry_sums})
industry_sums_total = industry_sums_total.sort_values(by='Total General', ascending=False)

# Mostrar la tabla
industry_sums_total
```

```python
# Agrupar y sumar los valores por categoría de habilidades
industry_sums = df_ind_mig_ar_melted.groupby('industry_name')['Value'].sum()

# Crear un DataFrame para mostrar el total general
industry_sums_total = pd.DataFrame({'Total General': industry_sums})

# Ordenar en orden descendente por 'Total General' para encontrar los 10 mayores
top_10 = industry_sums_total.nlargest(10, 'Total General')

# Ordenar en orden ascendente por 'Total General' para encontrar los 10 menores
bottom_10 = industry_sums_total.nsmallest(10, 'Total General')

# Crear un gráfico de barras horizontales para los 10 mayores
plt.figure(figsize=(10, 6))
plt.barh(top_10.index, top_10['Total General'], color='blue')

# Configurar el gráfico de los 10 mayores
plt.title('Top 10 industry_name con mayor valor')
plt.xlabel('Total General')
plt.ylabel('industry_name')

# Mostrar el gráfico de los 10 mayores
plt.show()

# Crear un gráfico de barras horizontales para los 10 menores
plt.figure(figsize=(10, 6))
plt.barh(bottom_10.index, bottom_10['Total General'], color='red')

# Configurar el gráfico de los 10 menores
plt.title('Top 10 industry_name con menor valor')
plt.xlabel('Total General')
plt.ylabel('industry_name')

# Mostrar el gráfico de los 10 menores
plt.show()
```

Impacto de las Tendencias Económicas:

En resumen, la migración neta de talento tiene un impacto significativo en diferentes industrias. Algunas industrias experimentan migración neta positiva, lo que indica un aumento en la atracción de talento, mientras que otras enfrentan migración neta negativa, lo que podría requerir medidas para atraer y retener talento en esas áreas. Las tendencias económicas y el desempeño de la industria pueden influir en estos patrones.
