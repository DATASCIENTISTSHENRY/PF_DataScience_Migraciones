```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leo el dataset
df_dest = pd.read_excel('/Users/gaston/Downloads/undesa_pd_2020_ims_stock_by_sex_destination_and_origin.xlsx',sheet_name = 'Table 1')
df_dest.head()
```

```python
#Borro las primeras filas del df
df_dest = df_dest.iloc[9:]
df_dest.head()

# Restablezco los índices del DataFrame después de eliminar la primera fila
df_dest = df_dest.reset_index(drop=True)

# Establezco la primera fila como los nombres de las columnas
df_dest.columns = df_dest.iloc[0]

# Elimino la primera fila actual
df_dest = df_dest.iloc[1:]

# Restablezco los índices del DataFrame después de eliminar la primera fila
df_dest = df_dest.reset_index(drop=True)
df_dest.head()
```

```python
#Mantengo solo las primeras 14 columnas ya que tienen la info con los totales para ambos sexos
df_dest = df_dest.iloc[:, :14]
df_dest
```

```python
# Elimino los espacios en blanco al principio de la columna 'Region, development group, country or area of origin'
df_dest['Region, development group, country or area of origin'] = df_dest['Region, development group, country or area of origin'].str.lstrip()
df_dest.head()
```

```python
#Defino un dataframe solo con Argentina como origen
df_dest_ar = df_dest.loc[df_dest['Region, development group, country or area of origin']=='Argentina']
df_dest_ar = df_dest_ar.reset_index(drop=True)
df_dest_ar.head()
```

```python
# Filtro los valores con 3 espacios al principio ya que en el dataset agrupa por región y los países están con 3 espacios al principio
# entonces para quedarme solo con los países y no sumar los valores más de una vez lo hago así
filtro = df_dest_ar['Region, development group, country or area of destination'].str.startswith('   ')
resultado = df_dest_ar[filtro]
df_dest_ar_f = resultado
df_dest_ar_f
```

```python
# Convierto los nombres de las columnas a cadenas (strings)
df_dest_ar_f.columns = df_dest_ar_f.columns.astype(str)
df_dest_ar_f

# Elimin0 ".0" de los nombres de las columnas
df_dest_ar_f.columns = df_dest_ar_f.columns.str.replace('.0', '')
df_dest_ar_f.head()
```

```python
#reseteo los índices
df_dest_ar_f = df_dest_ar_f.reset_index(drop=True)
df_dest_ar_f
```

```python
# Convierto las columnas de años a tipo numérico
years_columns = df_dest_ar_f.columns[7:]
df_dest_ar_f[years_columns] = df_dest_ar_f[years_columns].apply(pd.to_numeric, errors='coerce')

# Agrupo y sumo los valores para cada destino a lo largo de todos los años
total_incidences = df_dest_ar_f.groupby('Region, development group, country or area of destination')[years_columns].sum()

# Calculo la suma total para cada destino
total_incidences['Total'] = total_incidences.sum(axis=1)

# Selecciono los 10 destinos con los valores más altos
top_10_destinations = total_incidences.nlargest(10, 'Total')

# Muestro solamente el total
top_10_destinations = top_10_destinations[['Total']]

# Imprimo el resultado
print("Top 10 destinos con la mayor incidencia acumulada a lo largo de todos los años:\n")
print(top_10_destinations)

```

```python
# Elimino los espacios en blanco al principio de la columna 'Region, development group, country or area of origin'
df_dest_ar_f['Region, development group, country or area of destination'] = df_dest_ar_f['Region, development group, country or area of destination'].str.lstrip()
df_dest_ar_f.head()
```
